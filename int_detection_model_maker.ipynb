```
import numpy as np

from tflite_model_maker import model_spec
from tflite_model_maker import text_classifier
from tflite_model_maker.config import ExportFormat
from tflite_model_maker.text_classifier import AverageWordVecSpec
from tflite_model_maker.text_classifier import DataLoader

import tensorflow as tf

tf.get_logger().setLevel('ERROR')
```

### Average word vec

```
spec = model_spec.get('average_word_vec')
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/data/elma/df_group.csv")
df
```

```
train_data = DataLoader.from_csv(
      filename='/content/drive/MyDrive/data/elma/train.csv',
      text_column='phrase',
      label_column='intent',
      model_spec=spec,
      is_training=True)

val_data = DataLoader.from_csv(
      filename='/content/drive/MyDrive/data/elma/val.csv',
      text_column='phrase',
      label_column='intent',
      model_spec=spec,
      is_training=False)

test_data = DataLoader.from_csv(
      filename='/content/drive/MyDrive/data/elma/test.csv',
      text_column='phrase',
      label_column='intent',
      model_spec=spec,
      is_training=False)
```

```
model = text_classifier.create(train_data, 
                               model_spec=spec, 
                               validation_data=val_data, 
                               epochs=1000)
```

```
loss, acc = model.evaluate(test_data)
```
